```python
# Loading the flight dataset
import seaborn as sns
import matplotlib.pyplot as plt


flights = sns.load_dataset("flights") 
# display the few rown
print(flights.head())
```

```python

# Scatter plot
sns.relplot(
    data=flights, x="month", y="passengers", hue="year", kind="scatter"
)

plt.show()
```

```python
# line plot
sns.relplot(
    data=flights, x="month", y="passengers", hue="year", kind="line"
)

plt.show()
```

```python
# bar plot
sns.barplot(data=flights, x="month", y="passengers")
plt.xticks(rotation=45)

plt.show()
```

```python
# box plot
sns.boxplot(data=flights, x="month", y="passengers")
plt.xticks(rotation=45)

plt.show()
```

```python
#swarmplot
sns.swarmplot(data=flights, x="month", y="passengers")
plt.xticks(rotation=45)

plt.show()
```

```python
#displot
sns.displot(data=flights, x="month", y="passengers")
plt.xticks(rotation=45)

plt.show()
```

```python
#hisplot

sns.histplot(data=flights, x="passengers", bins=20)

plt.show()
```

```python
tips = sns.load_dataset("tips")
sns.pairplot(tips)
plt.show()
```
